Описать функцию ShiftRight3(A, B, C), выполняющую правый циклический сдвиг: значение A переходит в B, значение B — в C, значение C — в A (A, B, C — вещественные параметры, являющиеся одновременно входными и выходными). С помощью этой функции выполнить правый циклический сдвиг для двух данных наборов из трех чисел: (A1, B1, C1) и (A2, B2, C2).

```python
def print_symbols(count, symbol='*'):
    """
    Функция выводит строку, содержащую заданное число символов

    Параметры:
    count - количество символов для вывода
    symbol - символ для вывода (по умолчанию '*')
    """
    try:

        count = int(count)
        if count <= 0:
            print("Ошибка: количество символов должно быть больше 0!")
            return

        result = symbol * count
        print(f"Результат: {result}")

    except ValueError:
        print("Ошибка: введите целое число!")
    except Exception as e:
        print(f"Произошла ошибка: {e}")
print("=== Программа для вывода строки с заданным числом символов ===")

while True:
    try:

        num = input("Введите количество символов: ")

        sym = input("Введите символ для вывода (по умолчанию '*'): ")

        if sym == '':
            sym = '*'
        elif len(sym) > 1:
            print("Внимание: введено более одного символа, будет использован первый символ")
            sym = sym[0]

        print_symbols(num, sym)

        choice = input("\nХотите продолжить? (да/нет): ").lower()
        if choice not in ['да', 'yes', 'y', 'д']:
            break

    except KeyboardInterrupt:
        print("\nПрограмма завершена пользователем")
        break
    except Exception as e:
        print(f"Неожиданная ошибка: {e}")
        break

print("\nПрограмма завершена!")
```

```python
def ShiftRight3(A, B, C):
    """
    Функция выполняет правый циклический сдвиг:
    A → B, B → C, C → A

    Параметры (входные и выходные):
    A, B, C - вещественные числа
    """

    temp = C

    C = B
    B = A
    A = temp

    return A, B, C

print("=== Программа для правого циклического сдвига трех чисел ===")

def process_set(set_number):
    """
    Функция для обработки одного набора чисел
    """
    print(f"\n--- Набор {set_number} ---")

    while True:
        try:
            A = float(input("Введите число A: "))
            B = float(input("Введите число B: "))
            C = float(input("Введите число C: "))
            break
        except ValueError:
            print("Ошибка: введите числа правильно!")

    print(f"Исходные значения: A={A}, B={B}, C={C}")

    A, B, C = ShiftRight3(A, B, C)

    print(f"После сдвига:     A={A}, B={B}, C={C}")

    return A, B, C

print("Введите данные для первого набора:")
result1 = process_set(1)

print("\nВведите данные для второго набора:")
result2 = process_set(2)

print("\n=== Дополнительная демонстрация ===")
print("Выполним 3 правых циклических сдвига подряд для первого набора:")

print("\nВведите исходные значения для демонстрации:")
try:
    demo_A = float(input("Введите число A: "))
    demo_B = float(input("Введите число B: "))
    demo_C = float(input("Введите число C: "))

    print(f"\nИсходные значения: A={demo_A}, B={demo_B}, C={demo_C}")

    for i in range(1, 4):
        demo_A, demo_B, demo_C = ShiftRight3(demo_A, demo_B, demo_C)
        print(f"После {i} сдвига:  A={demo_A}, B={demo_B}, C={demo_C}")

except ValueError:
    print("Ошибка ввода для демонстрации")

print("\nПрограмма завершена!")
```
