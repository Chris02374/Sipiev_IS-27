Практическая Работа 4
Вариант 33:
1. Дано вещественное число A и целое число N (>0). Используя один цикл, найти сумму 1 + A + A2 + A3 + ... + AN.
2. Дано целое число N (> 0). Используя операции деления нацело и взятия остатка от деления, найти число, полученное при прочтении числа N справа налево.

```python
A = input("Введите вещественное число A: ")
N = input("Введите целое число N (>0): ")
while True:
    try:
        A = float(A)
        N = int(N)
        if N <= 0:
            print("Ошибка: N должно быть больше 0!")
            break
        total = 0
        power = 1
        i = 0
        while i <= N:
            total += power
            power *= A
            i += 1

        # Вывод результата
        print(f"Сумма 1 + A + A^2 + ... + A^{N} = {total}")
        print("\nАльтернативное решение с циклом for:")
        total2 = 0
        for i in range(N + 1):
            total2 += A ** i
        print(f"Сумма 1 + A + A^2 + ... + A^{N} = {total2}")

        break

    except ValueError:
        print("Ошибка: неправильный формат чисел!")
        break
```

```python
N = input("Введите целое число N (>0): ")
while True:
    try:
        N = int(N)
        if N <= 0:
            print("Ошибка: N должно быть больше 0!")
            break
        original = N
        reversed_number = 0

        print("\nПроцесс переворота числа:")
        while N > 0:
            last_digit = N % 10
            print(f"Текущая цифра: {last_digit}")
            reversed_number = reversed_number * 10 + last_digit
            N = N // 10
            print(f"Оставшееся число: {N}")
        print(f"\nИсходное число: {original}")
        print(f"Число справа налево: {reversed_number}")
        print(f"Проверка (строковый метод): {str(original)[::-1]}")

        break

    except ValueError:
        print("Ошибка: введите целое число!")
        break
```
