 1. Даны координаты двух различных полей шахматной доски х1, у1, х2, у2 (целые числа, лежащие в диапазоне 1-8). Проверить истинность высказывания: «Данные поля имеют одинаковый цвет».
 2. Даны три числа. Найти наименьшее из них.

```python
x1 = input("Введите координату x1 (1-8): ")
y1 = input("Введите координату y1 (1-8): ")
x2 = input("Введите координату x2 (1-8): ")
y2 = input("Введите координату y2 (1-8): ")

while True:
    try:
        x1 = int(x1)
        y1 = int(y1)
        x2 = int(x2)
        y2 = int(y2)

        if not (1 <= x1 <= 8 and 1 <= y1 <= 8 and 1 <= x2 <= 8 and 1 <= y2 <= 8):
            print("Ошибка: координаты должны быть в диапазоне 1-8")
            break

        if x1 == x2 and y1 == y2:
            print("Ошибка: поля одинаковые")
            break
        color1 = (x1 + y1) % 2
        color2 = (x2 + y2) % 2

        if color1 == color2:
            print("Истина: поля имеют одинаковый цвет")
        else:
            print("Ложь: поля имеют разный цвет")

        break

    except ValueError:
        print("Ошибка: введите целые числа!")
        break
```

```python
a = input("Введите первое число: ")
b = input("Введите второе число: ")
c = input("Введите третье число: ")

while True:
    try:
        a = float(a)
        b = float(b)
        c = float(c)

        if a < b:
            min_number = a
        else:
            min_number = b
        if c < min_number:
            min_number = c

        print(f"Наименьшее число: {min_number}")

        break

    except ValueError:
        print("Ошибка: введите числа!")
        break
```
