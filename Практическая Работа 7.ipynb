Вариант 33:
 1. Даны строки S и S0. Удалить из строки S последнюю подстроку, совпадающую с S0. Если совпадающих подстрок нет, то вывести строку S без изменений.
 2. Дана строка, состоящая из русских слов, набранных заглавными буквами и разделенных пробелами (одним или несколькими). Найти количество слов, которые содержат хотя бы одну букву «А».

```python
# Программа удаляет из строки S последнюю подстроку, совпадающую с S0

print("=== Задание 1: Удаление последней подстроки ===")

# Ввод строки S
S = input("Введите основную строку S: ")

# Ввод подстроки S0
S0 = input("Введите подстроку S0 для удаления: ")

print(f"\nИсходная строка S: '{S}'")
print(f"Подстрока S0: '{S0}'")

# Проверяем, содержится ли S0 в S
if S0 in S:
    # Находим последнее вхождение S0 в S
    last_index = S.rfind(S0)

    # Удаляем последнее вхождение S0
    # Берем часть строки до последнего вхождения
    # и часть строки после последнего вхождения
    result = S[:last_index] + S[last_index + len(S0):]

    print(f"\nПодстрока S0 найдена в позиции {last_index}")
    print(f"Строка после удаления последнего вхождения S0: '{result}'")

    # Альтернативный способ для проверки
    print(f"\nПроверка (разделение строки):")
    print(f"  Часть до S0: '{S[:last_index]}'")
    print(f"  Часть после S0: '{S[last_index + len(S0):]}'")

else:
    print(f"\nПодстрока S0 не найдена в строке S")
    print(f"Строка без изменений: '{S}'")
    result = S

# Дополнительная проверка с несколькими вхождениями
print(f"\n=== Дополнительный анализ ===")
all_occurrences = []
start = 0
while True:
    idx = S.find(S0, start)
    if idx == -1:
        break
    all_occurrences.append(idx)
    start = idx + 1

if all_occurrences:
    print(f"Все вхождения S0 в S (позиции): {all_occurrences}")
    print(f"Количество вхождений: {len(all_occurrences)}")
else:
    print("Вхождения S0 в S отсутствуют")
```

```python
# Программа находит количество слов, которые содержат хотя бы одну букву «А»

print("\n=== Задание 2: Подсчет слов с буквой 'А' ===")

# Ввод строки
text = input("Введите строку (русские слова заглавными буквами, разделенные пробелами): ")

print(f"\nИсходная строка: '{text}'")

# Разбиваем строку на слова
# Убираем лишние пробелы с помощью split() без параметров
words = text.split()

print(f"\nСписок слов: {words}")
print(f"Количество слов: {len(words)}")

# Счетчик слов с буквой 'А'
count_with_A = 0
words_with_A = []

# Проверяем каждое слово
for word in words:
    # Проверяем, что слово состоит из русских букв (опционально)
    if 'А' in word:
        count_with_A += 1
        words_with_A.append(word)

print(f"\nСлова, содержащие букву 'А': {words_with_A}")
print(f"Количество слов с буквой 'А': {count_with_A}")

# Дополнительный анализ
print(f"\n=== Дополнительный анализ ===")

for i, word in enumerate(words):
    if 'А' in word:
        # Находим все позиции буквы 'А' в слове
        positions = [str(pos + 1) for pos, char in enumerate(word) if char == 'А']
        print(f"Слово '{word}' (№{i + 1}): содержит 'А' на позициях {', '.join(positions)}")
    else:
        print(f"Слово '{word}' (№{i + 1}): не содержит 'А'")

# Проверка на другие заглавные русские буквы (опционально)
print(f"\n=== Проверка регистра букв ===")
has_lowercase = False
for word in words:
    for char in word:
        if char.isalpha() and char.islower():
            has_lowercase = True
            break

if has_lowercase:
    print("Внимание: в строке обнаружены строчные буквы!")
else:
    print("Все буквы в словах заглавные (как и требуется)")

# Подсчет общей статистики по букве 'А'
total_A_count = sum(word.count('А') for word in words)
print(f"Общее количество букв 'А' во всех словах: {total_A_count}")
```
