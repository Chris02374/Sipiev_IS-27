Практическая работа 6
Вариант 33:
1. Дан список размера N (N — четное число). Поменять местами первую и вторую половины списка.

2. Дан список A размера N. Сформировать новый список B того же размера по следующему правилу: элемент BK равен среднему арифметическому элементов списка A с номерами от K до N.

3. Дан список размера N и целое число K (1 < K < N). Осуществить сдвиг элементов списка влево на K позиций (при этом AN перейдет в AN-K, AN-1 — в AN-K-1, ..AK+1 — в A1, а исходное значение K первых элементов будет потеряно). Последние K элементов полученного списка положить равными 0.


```python
# Программа меняет местами первую и вторую половины списка

print("=== Задание 1: Поменять местами первую и вторую половины списка ===")

# Ввод размера списка с обработкой исключений
while True:
    try:
        N = int(input("Введите размер списка N (четное число): "))
        if N <= 0:
            print("Ошибка: размер списка должен быть больше 0!")
        elif N % 2 != 0:
            print("Ошибка: размер списка должен быть четным числом!")
        else:
            break
    except ValueError:
        print("Ошибка: введите целое число!")

# Создаем список
A = []
print(f"\nВведите {N} элементов списка:")
for i in range(N):
    while True:
        try:
            value = int(input(f"Элемент A[{i}]: "))
            A.append(value)
            break
        except ValueError:
            print("Ошибка: введите целое число!")

print(f"\nИсходный список A: {A}")

# Меняем местами первую и вторую половины
half = N // 2
B = A[half:] + A[:half]

print(f"Список после обмена половин B: {B}")

# Альтернативный способ с использованием срезов в исходном списке
# Создаем копию исходного списка
C = A.copy()
C[:half], C[half:] = C[half:], C[:half]
print(f"Проверка (альтернативный метод): {C}")
```

```python
# Программа формирует список B по правилу: B[K] = среднее арифметическое элементов A с K до N

print("\n=== Задание 2: Формирование списка B по правилу ===")

# Ввод размера списка с обработкой исключений
while True:
    try:
        N = int(input("Введите размер списка N: "))
        if N <= 0:
            print("Ошибка: размер списка должен быть больше 0!")
        else:
            break
    except ValueError:
        print("Ошибка: введите целое число!")

# Создаем список A
A = []
print(f"\nВведите {N} элементов списка A:")
for i in range(N):
    while True:
        try:
            value = int(input(f"Элемент A[{i}]: "))
            A.append(value)
            break
        except ValueError:
            print("Ошибка: введите целое число!")

print(f"\nИсходный список A: {A}")

# Создаем список B
B = []

# Формируем список B по правилу
for K in range(N):
    # Вычисляем среднее арифметическое элементов с K до N
    if K < N:  # Проверка, что K не выходит за пределы
        elements = A[K:]  # Берем элементы с K до конца
        average = sum(elements) / len(elements)
        B.append(round(average, 2))  # Округляем до 2 знаков
    else:
        B.append(0)  # На случай, если K = N (но такого не должно быть)

print(f"Полученный список B: {B}")

# Дополнительный вывод для наглядности
print("\nПодробный расчет:")
for K in range(N):
    elements = A[K:]
    avg = sum(elements) / len(elements)
    print(f"B[{K}] = среднее({A[K:]}) = {round(avg, 2)}")
```

```python
# Программа осуществляет сдвиг элементов списка влево на K позиций

print("\n=== Задание 3: Сдвиг элементов списка влево на K позиций ===")

# Ввод размера списка с обработкой исключений
while True:
    try:
        N = int(input("Введите размер списка N: "))
        if N <= 0:
            print("Ошибка: размер списка должен быть больше 0!")
        else:
            break
    except ValueError:
        print("Ошибка: введите целое число!")

# Создаем список A
A = []
print(f"\nВведите {N} элементов списка A:")
for i in range(N):
    while True:
        try:
            value = int(input(f"Элемент A[{i}]: "))
            A.append(value)
            break
        except ValueError:
            print("Ошибка: введите целое число!")

print(f"\nИсходный список A: {A}")

# Ввод K с обработкой исключений
while True:
    try:
        K = int(input(f"\nВведите число K (1 < K < {N}): "))
        if K <= 1 or K >= N:
            print(f"Ошибка: K должно быть больше 1 и меньше {N}!")
        else:
            break
    except ValueError:
        print("Ошибка: введите целое число!")

# Выполняем сдвиг влево на K позиций
# Создаем копию исходного списка
B = A.copy()

# Сдвигаем элементы влево
for i in range(N - K):
    B[i] = B[i + K]

# Последние K элементов заполняем нулями
for i in range(N - K, N):
    B[i] = 0

print(f"\nСписок после сдвига на {K} позиций влево: {B}")

# Дополнительный вывод для наглядности
print("\nПроцесс сдвига:")
print("1. Перенос элементов:")
for i in range(N - K):
    print(f"   B[{i}] = A[{i + K}] = {A[i + K]}")

print(f"\n2. Заполнение последних {K} элементов нулями:")
for i in range(N - K, N):
    print(f"   B[{i}] = 0")

# Альтернативный способ с использованием срезов
print(f"\nПроверка (альтернативный метод со срезами):")
C = A[K:] + [0] * K
print(f"   A[K:] = {A[K:]}")
print(f"   [0]*K = {[0] * K}")
print(f"   Результат: {C}")
```
